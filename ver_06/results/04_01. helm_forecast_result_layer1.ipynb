```python
def total_result(acc_type, data_type, list_00, list_01, list_02, list_03):
    train_mean_00 = np.mean(list_00)
    train_std_00 = np.std(list_00)
    
    train_mean_01 = np.mean(list_01)
    train_std_01 = np.std(list_01)
    
    train_mean_02 = np.mean(list_02)
    train_std_02 = np.std(list_02)
    
    train_mean_03 = np.mean(list_03)
    train_std_03 = np.std(list_03)
    
    print data_type
    
    print 'Case 00 [' + acc_type + '/std] = ',
    print str(train_mean_00) + ' / ' + str(train_std_00)
    
    print 'Case 01 [' + acc_type + '/std] = ',
    print str(train_mean_01) + ' / ' + str(train_std_01)
    
    print 'Case 02 [' + acc_type + '/std] = ',
    print str(train_mean_02) + ' / ' + str(train_std_02)
    
    print 'Case 03 [' + acc_type + '/std] = ',
    print str(train_mean_03) + ' / ' + str(train_std_03)
```

```python
def total_RMSE_graph(acc_type, data_type, list_00, list_01, list_02, list_03):
    if data_type is 'Train':
        plot_train_00 = sns.distplot(list_00, norm_hist=True, kde_kws={"label":"case 00"}, color='#658cbb')
        plot_train_01 = sns.distplot(list_01, norm_hist=True, kde_kws={"label":"case 01"}, color='#7bb274')
        plot_train_02 = sns.distplot(list_02, norm_hist=True, kde_kws={"label":"case 02"}, color='#d3494e')
        plot_train_03 = sns.distplot(list_03, norm_hist=True, kde_kws={"label":"case 03"}, color='#916e99')
    elif data_type is 'Test':
        plot_test_00 = sns.distplot(list_00, norm_hist=True, kde_kws={"label":"case 00"}, color='blue')
        plot_test_01 = sns.distplot(list_01, norm_hist=True, kde_kws={"label":"case 01"}, color='green')
        plot_test_02 = sns.distplot(list_02, norm_hist=True, kde_kws={"label":"case 02"}, color='red')
        plot_test_03 = sns.distplot(list_03, norm_hist=True, kde_kws={"label":"case 03"}, color='purple')
    
    plt.title('H-ELM', fontsize=13)
    plt.xlabel(acc_type + ' ' + data_type, fontsize=10)
    plt.ylabel('Density', fontsize=10)

    plt.xlim(0.06, 0.17)
    plt.ylim(0, 5000)
```

```python
def total_MAPE_graph(acc_type, data_type, list_00, list_01, list_02, list_03):
    if data_type is 'Train':
        plot_train_00 = sns.distplot(list_00, norm_hist=True, kde_kws={"label":"case 00"}, color='#658cbb')
        plot_train_01 = sns.distplot(list_01, norm_hist=True, kde_kws={"label":"case 01"}, color='#7bb274')
        plot_train_02 = sns.distplot(list_02, norm_hist=True, kde_kws={"label":"case 02"}, color='#d3494e')
        plot_train_03 = sns.distplot(list_03, norm_hist=True, kde_kws={"label":"case 03"}, color='#916e99')
    elif data_type is 'Test':
        plot_test_00 = sns.distplot(list_00, norm_hist=True, kde_kws={"label":"case 00"}, color='blue')
        plot_test_01 = sns.distplot(list_01, norm_hist=True, kde_kws={"label":"case 01"}, color='green')
        plot_test_02 = sns.distplot(list_02, norm_hist=True, kde_kws={"label":"case 02"}, color='red')
        plot_test_03 = sns.distplot(list_03, norm_hist=True, kde_kws={"label":"case 03"}, color='purple')

    plt.title('H-ELM', fontsize=13)
    plt.xlabel(acc_type + ' ' + data_type, fontsize=10)
    plt.ylabel('Density', fontsize=10)

    plt.xlim(7, 30)
    plt.ylim(0, 27)
```

```python
def total_MAE_graph(acc_type, data_type, list_00, list_01, list_02, list_03):
    if data_type is 'Train':
        plot_train_00 = sns.distplot(list_00, norm_hist=True, kde_kws={"label":"case 00"}, color='#658cbb')
        plot_train_01 = sns.distplot(list_01, norm_hist=True, kde_kws={"label":"case 01"}, color='#7bb274')
        plot_train_02 = sns.distplot(list_02, norm_hist=True, kde_kws={"label":"case 02"}, color='#d3494e')
        plot_train_03 = sns.distplot(list_03, norm_hist=True, kde_kws={"label":"case 03"}, color='#916e99')
    elif data_type is 'Test':
        plot_test_00 = sns.distplot(list_00, norm_hist=True, kde_kws={"label":"case 00"}, color='blue')
        plot_test_01 = sns.distplot(list_01, norm_hist=True, kde_kws={"label":"case 01"}, color='green')
        plot_test_02 = sns.distplot(list_02, norm_hist=True, kde_kws={"label":"case 02"}, color='red')
        plot_test_03 = sns.distplot(list_03, norm_hist=True, kde_kws={"label":"case 03"}, color='purple')

    plt.title('H-ELM', fontsize=13)
    plt.xlabel(acc_type + ' ' + data_type, fontsize=10)
    plt.ylabel('Density', fontsize=10)

    plt.xlim(7000, 22000)
    plt.ylim(0, 0.03)
```
