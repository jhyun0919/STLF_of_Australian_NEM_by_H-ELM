```python
MAE_00 = ap.Numpy2Pandas.add_number_index(ap.Numpy2Pandas.mae_data2df(np.array(f_00['MAE'].tolist())))
MAE_01 = ap.Numpy2Pandas.add_number_index(ap.Numpy2Pandas.mae_data2df(np.array(f_01['MAE'].tolist())))
MAE_02 = ap.Numpy2Pandas.add_number_index(ap.Numpy2Pandas.mae_data2df(np.array(f_02['MAE'].tolist())))
MAE_03 = ap.Numpy2Pandas.add_number_index(ap.Numpy2Pandas.mae_data2df(np.array(f_03['MAE'].tolist())))
```

```python
df_00 = ap.Numpy2Pandas.add_number_index(ap.Numpy2Pandas.acc_data2df(np.array(RMSE_00)))
df_01 = ap.Numpy2Pandas.add_number_index(ap.Numpy2Pandas.acc_data2df(np.array(RMSE_01)))
df_02 = ap.Numpy2Pandas.add_number_index(ap.Numpy2Pandas.acc_data2df(np.array(RMSE_02)))
df_03 = ap.Numpy2Pandas.add_number_index(ap.Numpy2Pandas.acc_data2df(np.array(RMSE_03)))
```

---

```python
def total_result(acc_type, df_00, df_01, df_02, df_03):
    idx_00 = df_00[acc_type].idxmin()-1
    idx_01 = df_01[acc_type].idxmin()-1
    idx_02 = df_02[acc_type].idxmin()-1
    idx_03 = df_03[acc_type].idxmin()-1
    
    min_00 = df_00[acc_type].iloc[idx_00]
    min_01 = df_00[acc_type].iloc[idx_01]
    min_02 = df_00[acc_type].iloc[idx_02]
    min_03 = df_00[acc_type].iloc[idx_03]
    
    print 'case_00 : idx = ' + str(idx_00) + ' / value = ' + str(min_00)
    print 'case_01 : idx = ' + str(idx_01) + ' / value = ' + str(min_01)
    print 'case_02 : idx = ' + str(idx_02) + ' / value = ' + str(min_02)
    print 'case_03 : idx = ' + str(idx_03) + ' / value = ' + str(min_03)
```

```python
def total_RMSE_graph(df_00, df_01, df_02, df_03):
    font_size = 9

    plt.subplot(221)
    plt.plot(df_00[:], color='blue')
    plt.title('case 00', fontsize=font_size)

    plt.subplot(222)
    plt.plot(df_01[30:], color='green')
    plt.ylim(0.077, 0.1)
    plt.title('case 01', fontsize=font_size)

    plt.subplot(223)
    plt.plot(df_02[30:], color='red')
    plt.ylim(0.077, 0.1)
    plt.title('case 02', fontsize=font_size)

    plt.subplot(224)
    plt.plot(df_03[30:], color='purple')
    plt.ylim(0.077, 0.1)
    plt.title('case 03', fontsize=font_size)
```

```python
def total_MAPE_graph(df_00, df_01, df_02, df_03):
    font_size = 9

    plt.subplot(221)
    plt.plot(df_00[:], color='blue')
    plt.title('case 00', fontsize=font_size)

    plt.subplot(222)
    plt.plot(df_01[30:], color='green')
    plt.ylim(11, 15)
    plt.title('case 01', fontsize=font_size)

    plt.subplot(223)
    plt.plot(df_02[30:], color='red')
    plt.ylim(11, 15)
    plt.title('case 02', fontsize=font_size)

    plt.subplot(224)
    plt.plot(df_03[30:], color='purple')
    plt.ylim(11, 15)
    plt.title('case 03', fontsize=font_size)
```

```python
def total_MAE_graph(df_00, df_01, df_02, df_03):
    font_size = 9

    plt.subplot(221)
    plt.plot(df_00[:], color='blue')
    plt.title('case 00', fontsize=font_size)

    plt.subplot(222)
    plt.plot(df_01[30:], color='green')
    plt.ylim(10000, 13000)
    plt.title('case 01', fontsize=font_size)

    plt.subplot(223)
    plt.plot(df_02[30:], color='red')
    plt.ylim(10000, 13000)
    plt.title('case 02', fontsize=font_size)

    plt.subplot(224)
    plt.plot(df_03[30:], color='purple')
    plt.ylim(10000, 13000)
    plt.title('case 03', fontsize=font_size)
```
